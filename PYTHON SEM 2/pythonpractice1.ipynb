```python
print("Enter a number")
number = input("number")
number1 = int(number)
if number1 == 0:
    print("the number is 0")
elif(number1 > 0):
    print("The number is positive")
else:
    print("The number is negative")
if number1%2 == 0:
    print("The number is even")
else:
    print("The number is odd")
```
