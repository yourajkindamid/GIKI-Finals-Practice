```python
d1 = {'a': 2, 'b': 5}
d2 = {'b': 3, 'c': 4}
d3 = {}
d3.update(d1)
for key in d2:
    if key in d3:
        d3[key] += d2[key]
    else:
        d3[key] = d2[key]
print(d3)         
```

```python
d = {'a': 1, 'b': 4, 'c': 4, 'd': 2, 'e': 5}

largest = {}  # Dictionary to store the largest element
second_largest = {}  # Dictionary to store the second largest element
lrg = 0  # Largest value
second_lrg = 0  # Second largest value

for x in d:
    if d[x] > lrg:  
        second_lrg = lrg  # Update second largest before changing largest
        second_largest = largest.copy()  # Store previous largest entry
        lrg = d[x]  # Update largest value
        largest = {x: d[x]}  # Store new largest key-value pair
    elif d[x] > second_lrg and d[x] < lrg:
        second_lrg = d[x]  # Update second largest value
        second_largest = {x: d[x]}  # Store new second largest key-value pair

print("Largest:", largest)
print("Second Largest:", second_largest)
  
```
