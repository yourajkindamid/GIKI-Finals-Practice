```python
class Book:
    def __init__ (self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages
    def get_details(self):
        return f"Title:{self.title}\nAuthor:{self.author}\nPages:{self.pages}"
    def is_long(self):
        if int(self.pages) > 300:
            print("The book is long")
        else:
            print("The book is short")
obj = Book("youraj", "ali", 500)
print(obj.get_details())
obj.is_long()
```

```python
class Person:
    def __init__ (self, name, age):
        self.__name = name
        self.__age = age
    def set_name(self, name):
        self.__name = name
    def get_name(self):
        return self.__name
    def set_age(self, age):
        self.__age = age
    def get_age(self):
        return self.__age
obj = Person("Youraj", 19)
obj.set_name("Youraj")
print(obj.get_name())
obj.set_age(19)
print(obj.get_age())
```

```python
class Vehicle:
    def __init__ (self, brand, model):
        self.brand = brand
        self.model = model
    def details(self):
        return f"Model:{self.brand},{self.model}"
class Car(Vehicle):
    def __init__ (self, brand, model, seats):
        super().__init__(brand, model)
        self.seats = seats
    def details(self):
        return f"Car:{self.brand},{self.model}, Seats: {self.seats}"
class Bike(Vehicle):
    def __init__ (self, brand, model, type):
        super().__init__(brand, model)
        self.type = type
    def details(self):
       return f"Bike:{self.brand},{self.model}, Type:{self.type}"

obj1 = Car("Toyota", "Corolla", 3)
obj1.details()
```

```python
class BankAccount:
    def __init__ (self, account_number, balance):
        self.account_number = account_number
        self.balance = balance
    def __add__ (self, other):
        return self.balance + other.balance
acc1 = BankAccount(101, 5000)
acc2 = BankAccount(102, 3000)

acc3 = acc1 + acc2
print(acc3)  
```

```python
class CompoundInterest:
    def __init__ (self, principal, rate, times_compounded, time):
        self.principal = principal
        self.rate = rate
        self.times_compounded = times_compounded
        self.time = time
    def __pow__ (self, other=0):
        return self.principal * ((1+(self.rate/self.times_compounded))**(self.times_compounded*self.time))
    def __str__(self, other =0):
        return f"result = {self ** None}"
obj1 = CompoundInterest(principal=1000, rate=0.05, times_compounded=4, time=2)
print(obj1)
```
