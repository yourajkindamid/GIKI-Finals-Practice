```python
#TASK 1
temp = input("What is the temperature?")
tempint = int(temp)

if tempint < 0:
    print("Freezing")
elif tempint == 0 and tempint < 16:
    print("Cold")
elif tempint > 15 and tempint < 26:
   print("Mild")
else:
    print("Hot")

if tempint == 100:
        print("Boiling Point")
```
