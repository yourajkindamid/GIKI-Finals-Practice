```python
def find_max(*args):
    large = 0
    for x in args:
        if x > large:
            large = x
    return large
print(find_max(4, 7, 1, 9, 3))
```

```python
def multiply(a, b, c):
    return a*b*c
numbers = [2, 3, 4]
print(multiply(*numbers))
```

```python
def print_user(**kwargs):
    for key, value in kwargs.items():
        print(f"{key.capitalize()}:{value}")
print_user(name="Alice", age=25, city="NY")
```
