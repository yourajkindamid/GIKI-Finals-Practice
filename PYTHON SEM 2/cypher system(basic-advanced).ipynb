```python
text = 'youraj'
alpha = 'abcdefghijklmnopqrstuvwxyzabcdefghijklmnopqrstuvwxyz'
shift = 3
for encry in text:
    index = alpha.find(encry)
    shifted = index + shift
    print('og: ', alpha[index], ' ', 'encrypted: ', alpha[shifted])    
```

```python
text = 'Hello World'
alpha = 'abcdefghijklmnopqrstuvwxyz'
shift = 3
encrypted_text = ''
for tex in text.lower():
    index = alpha.find(tex)
    shifted_index = index + shift
    encrypted_text += alpha[shifted_index]
    print('original: ', tex, ' ', 'encrypted: ', encrypted_text)
```
