```python
#TASK1
import numpy as np
import matplotlib.pyplot as plt
names = np.array([("Student 1"), ("Student 2"), ("Student 3")])
marks = np.array([(30, 12, 39), (35, 40, 30), (25, 40, 20)])
subjects = np.array([("Stats"), ("English"), ("CompSci")])
sumstudmarks = marks.sum(axis=1)
print(f"Sum of Each Students Marks: {sumstudmarks}")
sum_students = []
for x in range(0, 3):
    y = sumstudmarks[x]
    sum_students.append(y)

avgstud = sumstudmarks/3
print(f"Average Marks of Each Student: {avgstud}")
sumsubmarks = marks.sum(axis = 0)
print(f"Sum of Marks for Each Subject: {sumsubmarks}")
avgsub = sumsubmarks/3
print(f"Average Marks of Subjects: {avgsub}")
stumaxmarks = marks.max(axis=1)
print(f"Maximum Marks of Students: {stumaxmarks}")
submaxmarks = marks.max(axis = 0)
print(f"Maximum Marks in Each Subject: {stumaxmarks}")
stuminmarks = marks.min(axis=1)
print(f"Minimum Marks of Students: {stuminmarks}")
subminmarks = marks.min(axis=0)
print(f"Minimum Marks of Students: {subminmarks}")

plt.bar(names, sum_students, color = 'b')
plt.xlabel('Student Names')
plt.ylabel('Student Marks')
plt.title('Sum of Student Marks')
plt.ioff()
plt.show()

for x in range(0, 3):
    plt.plot(subjects, marks[x], label = f"Student {1 + x}")
plt.ylabel('Per Subject Marks')
plt.xlabel('Subject Names')
plt.legend()
plt.show()
```

```python
#TASK 2
import numpy as np
import matplotlib.pyplot as plt

mean = 70
std_deviation = 10
datapoints = np.random.normal(70, 10, 1000)
calc_mean = np.mean(datapoints)
print(f"Calculated Mean = {calc_mean}")
calc_stddv = np.std(datapoints)
print(f"Calculated Standard Deviation = {calc_stddv}")
plt.hist(datapoints, bins = 10, histtype = 'bar', rwidth = 0.8)
plt.xlabel('Normal Distribution')
plt.ylabel('Marks')
plt.show()
plt.ioff()

study_hours = np.random.normal(15,2, 1000)
plt.scatter(study_hours,datapoints,label='Scatter', color = 'k', marker = 'o', s=15)
plt.xlabel("Study Hours")
plt.ylabel("Test Scores")
plt.title("Relation between Study Hours and Scores")
plt.show()
plt.ioff()
```

```python
#TASK3
import numpy as np
import matplotlib.pyplot as plt

age_groups = np.array([('0-10'), ('11-20'), ('21-30'), ('31-40'), ('41-50'), ('51-60'), ('61-70')])
population = np.array([(1000), (2000), (3000), (2500), (1500), (900), (500)])

plt.bar(age_groups, population, color = 'y')
plt.xlabel('Age Groups')
plt.ylabel('Population')
plt.show()
plt.ioff()

plt.barh(age_groups, population, color = 'r')
plt.xlabel('Population')
plt.ylabel('Age Groups')
plt.show()
plt.ioff()

colors = ['r', 'b', 'g', 'y', 'orange', 'b', 'c']
fig1, ax1 = plt.subplots()
ax1.pie(population, colors = colors, labels = age_groups, autopct = '%1.1f%%', startangle=90)

center_circle = plt.Circle((0, 0), .5, fc = 'white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)

ax1.axis('equal')
plt.tight_layout()
plt.title("Population Distribution through Pie Chart")
plt.show()
plt.ioff()


Men = np.array([(500), (1200), (1600), (1200), (700), (400), (150)])
Women = np.array([(500), (800), (1400), (1300), (800), (500), (350)])

plt.bar(age_groups, Men, color = 'b', label = "Men")
plt.bar(age_groups, Women, color = 'r', bottom = Men, label = "Women")
plt.legend()
plt.title("Distribution of Men and Women among population")
plt.show()
plt.ioff()
```
